```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

data = pd.read_csv("dataFiltrada.csv", low_memory = False)
data = data.loc[data['price'] < 15000000,:].drop('Unnamed: 0', axis = 1)
data['rango'] = data['price'].apply(lambda x: int(x/20000) + 1)
data['created_on'] = pd.to_datetime(data['created_on'])
data.info()
```

```python
dataPrecio = data.groupby('price_usd_per_m2').count()
```

```python
dataGroup = data.groupby('rango', as_index=False)['distancia_al_centro'].agg({'cantidad':'count'})
dataGroup = dataGroup.loc[dataGroup['cantidad'] > 10 ,:]
dataFinal = pd.merge(data, dataGroup, on= 'rango')
dataGroup.sort_values('rango')
```

```python
def tipoDePropiedad(x):
    tipos = {
        "apartment" : 1,
        "house" : 2,
        "PH" : 3,
        "store" : 4
    }
    return tipos[x] if x in tipos else x


def fechaNumerica(x):
    return 10000*x.year + 100*x.month + x.day
```

```python
dataFinal.loc[:, 'property_type'] = dataFinal.property_type.apply(tipoDePropiedad)
#dataFinal.loc[:, 'distancia_al_centro'] = dataFinal.distancia_al_centro.apply(lambda x: x/1000)
dataFinal.loc[:, 'created_on'] = dataFinal.created_on.apply(fechaNumerica)
dataCapital = dataFinal.loc[dataFinal.state_name.str.contains('Capital Federal'), :]
```
