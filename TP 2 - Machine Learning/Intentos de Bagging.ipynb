```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
def generarMuestra(df):
    muestra = df.sample(frac = 0.8, replace = True)
  

    x_train = np.array(muestra.loc[:,['created_on', 'property_type', 'lat', 'lon', 'cochera', 'surface_total_in_m2',\
                       'surface_covered_in_m2', 'seguridad', 'aire acondicionado', 'patio', 'pileta', 'amenities', 'zona',\
                        'distanciaCentro', 'distanciaAIndoamericano', 'distanciaAFuerteApache', 'distanciaACarlosGardel',\
                        'subtes_cercanos', 'trenes_cercanos']])
    y_train = np.array(muestra.loc[:, 'price_aprox_usd'])

    return x_train, y_train

def generarTest(df):
    test = df.sample(frac = 0.2)
  

    x_test = np.array(test.loc[:,['created_on', 'property_type', 'lat', 'lon', 'cochera', 'surface_total_in_m2',\
                       'surface_covered_in_m2', 'seguridad', 'aire acondicionado', 'patio', 'pileta', 'amenities', 'zona',\
                        'distanciaCentro', 'distanciaAIndoamericano', 'distanciaAFuerteApache', 'distanciaACarlosGardel',\
                        'subtes_cercanos', 'trenes_cercanos']])
    y_test = np.array(test.loc[:, 'price_aprox_usd'])

    return x_test, y_test

def generarMuestra2(df):
    muestra = df.sample(frac = 0.8, replace = True)
  

    x_train = np.array(muestra.loc[:,['created_on', 'property_type', 'barrio', 'cochera', 'surface_total_in_m2',\
                       'surface_covered_in_m2', 'seguridad', 'aire acondicionado', 'patio', 'pileta', 'amenities', 'zona',\
                        'distanciaCentro', 'distanciaAIndoamericano', 'distanciaAFuerteApache', 'distanciaACarlosGardel',\
                        'subtes_cercanos', 'trenes_cercanos']])
    y_train = np.array(muestra.loc[:, 'price_aprox_usd'])

    return x_train, y_train

def generarTest2(df):
    test = df.sample(frac = 0.2)
  

    x_test = np.array(test.loc[:,['created_on', 'property_type', 'barrio', 'cochera', 'surface_total_in_m2',\
                       'surface_covered_in_m2', 'seguridad', 'aire acondicionado', 'patio', 'pileta', 'amenities', 'zona',\
                        'distanciaCentro', 'distanciaAIndoamericano', 'distanciaAFuerteApache', 'distanciaACarlosGardel',\
                        'subtes_cercanos', 'trenes_cercanos']])
    y_test = np.array(test.loc[:, 'price_aprox_usd'])

    return x_test, y_test
```

```python
def salvarSupTotal(reg):
    if reg['surface_total_in_m2'] == 0:
        reg['surface_total_in_m2'] = 100
    return reg['surface_total_in_m2']

def salvarSupCovered(reg):
    
    if reg['surface_covered_in_m2'] == 0:
        reg['surface_covered_in_m2'] = 100
    return reg['surface_covered_in_m2']
```

```python
setTesteo2 = pd.read_csv("setTesteo.csv", low_memory=False)
testId2 = setTesteo2['id']
setTesteo2['created_on'] = pd.to_datetime(setTesteo2['created_on'])
dicc_tipos_prop = {"apartment" : 1, "house" : 2, "PH" : 3, "store" : 4}
dicc_tipos_zona = {"Bs.As. G.B.A. Zona Norte": 1, "Bs.As. G.B.A. Zona Oeste": 2, "Bs.As. G.B.A. Zona Sur": 3, "Capital Federal": 4, "Buenos Aires Interior": 4 }
setTesteo2.loc[:, 'property_type'] = setTesteo2.property_type.apply(lambda x: dicc_tipos_prop.get(x))
setTesteo2.loc[:, 'zona'] = setTesteo2.zona.apply(lambda x: dicc_tipos_zona.get(x))
setTesteo2.loc[:, 'created_on'] = setTesteo2.created_on.apply(fechaNumerica)
setTesteo2['surface_total_in_m2'] = setTesteo2.apply(salvarSupTotal, axis=1)
setTesteo2['surface_covered_in_m2'] = setTesteo2.apply(salvarSupCovered, axis=1)
setTesteo2 = setTesteo2.drop('Unnamed: 0', axis = 1)
setTesteo2 = setTesteo2.drop('barrio', axis = 1)
setTesteo2 = setTesteo2.drop('id', axis = 1)
setTesteo2
```
