```python
#Funcion para calcular distancias entre coordenadas (lat, lon)
from math import sin,cos,sqrt,asin,pi

#Formula del Harvesine sacada de
#https://joseguerreroa.wordpress.com/2012/11/13/distancia-entre-dos-puntos-de-la-superficie-terrestre-mediante-la-formula-de-haversine-con-python/
#coord = (lat, long)
def distancia(coord1, coord2):
    rTierra = 6374000 #metros
    c = pi/180 #para pasar a raddianes
    d = 2*rTierra*asin(sqrt(sin(c*(coord2[0]-coord1[0])/2)**2 + cos(c*coord1[0])*cos(c*coord2[0])*sin(c*(coord2[1]-coord1[1])/2)**2))
    return d

#reg = created_on, place_name, lat, lon, price_usd_per_m2
def distanciaANuevasEstaciones(reg):
    created_on, place_name, lat, lon, price_usd_per_m2 = reg
    geoPos = lat, lon
    #Ubicacion de las estaciones
    Cordoba = (-34.598691, -58.403841) #Inicio Obras: 17 ene 2012 - Inauguracion: 18 dic 2015
    SantaFe = (-34.594848, -58.402517) #Inicio Obras: 17 ene 2012 - Inauguracion: 12 jul 2016
    LasHeras = (-34.587879, -58.397970) #Inicio Obras: 17 ene 2012 - Inauguracion: 18 dic 2015

    distACordoba = distancia(geoPos, Cordoba)
    distASantaFe = distancia(geoPos, SantaFe)
    distALasHeras = distancia(geoPos, LasHeras)
    

    return min(distACordoba, distASantaFe, distALasHeras)

#calcula la distancia en metros entre la propiedad y la villa 1-11-14. Recibe por parametro un registro
def distanciaA1_11_14(reg):
    created_on, property_type, place_name, lat, lon, price_usd_per_m2, surface_total_in_m2 = reg
    geoPos = lat, lon
    #Ubicacion
    villa_1_11_14 = (-34.648784, -58.440711) #Tomamos este como el punto central de la villa

    distA1_11_14 = distancia(geoPos, villa_1_11_14)   

    return distA1_11_14

#calcula la distancia en metros entre la propiedad y la villa Fuerte Apacha. Recibe por parametro un registro
def distanciaAFuerteApache(reg):
    created_on, property_type, place_name, lat, lon, price_usd_per_m2, surface_total_in_m2, distanciaA1_11_14 = reg
    geoPos = lat, lon
    #Ubicacion
    fuerteApache = (-34.622374, -58.539921) #Tomamos este como el punto central de la villa

    distAfuerteApache = distancia(geoPos, fuerteApache)   

    return distAfuerteApache

distancia([-34.585149, -58.392506], (-34.587879, -58.397970))
```

```python
#elimino propiedades que no tengan ni latitud ni longitud
dataLatLon = dataBarriosNoNulos.loc[dataBarriosNoNulos.lat.notnull() & dataBarriosNoNulos.lon.notnull(), ['created_on', 'property_type', 'place_name', 'lat', 'lon', 'price_usd_per_m2', 'surface_total_in_m2']]

#agrego la distancia de las propiedades a las 2 villas que utilizaremos como ejemplo
dataLatLon.loc[:, 'distanciaA1_11_14'] = dataLatLon.apply(distanciaA1_11_14, axis = 1)
dataLatLon.loc[:, 'distanciaAfuerteApache'] = dataLatLon.apply(distanciaAFuerteApache, axis = 1)
dataLatLon = dataLatLon.sort_values(('distanciaA1_11_14'), ascending=True)

#elimino propiedades con las mismas latitud y longitud para no tener repetidas
dataLatLon = dataLatLon.drop_duplicates(['lat', 'lon'], keep='first')

#vamos a comparar propiedades que estan entre 1000 y 1500mts de distancia con propiedades entre 1500 y 2000mts de distancia
cercanasA1_11_14 = dataLatLon.loc[(dataLatLon['distanciaA1_11_14'] > 1000) & (dataLatLon['distanciaA1_11_14'] < 2000), :]
cercanasA1_11_14.loc[:, 'price'] = cercanasA1_11_14['price_usd_per_m2'] * cercanasA1_11_14['surface_total_in_m2']
cercanasA1_11_14.head()
```
