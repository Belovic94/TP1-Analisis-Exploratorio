```python
import datetime as datetime
import pandas as pd
import numpy as np
#import seaborn as sns
import matplotlib.pyplot as plt

#Permite ver los graficos sin colocar plt.show
%matplotlib inline 

plt.style.use('default')
#plt.rcParams['figure.figsize'] = (15, 5)

data = pd.read_csv("properati-AR-2017-08-01-properties-sell.csv")
data.head()
```

```python
dataCapGba = data.loc[(data.state_name.str.contains('Capital Federal') | data.state_name.str.contains('G.B.A') ) & data.operation.str.contains('sell') , ['created_on','operation', 'property_type', 'state_name', 'place_name', 'lat', 'lon', 'price', 'currency', 'surface_total_in_m2', 'surface_covered_in_m2', 'price_usd_per_m2', 'price_per_m2', 'floor', 'rooms', 'expenses', 'description']]
dataCapGba.sort_values('price_usd_per_m2', ascending = False)
```

```python
dataCapGbaMenoresA10K = dataCapGba.loc[dataCapGba['price_usd_per_m2'] < 10000,:]
dataCapGbamayoresA1000 = dataCapGbaMenoresA10K.loc[dataCapGbaMenoresA10K['price_usd_per_m2'] > 1000,:]
dataCapGbamayoresA1000.sort_values('price_usd_per_m2', ascending = False)
```

```python
dataCapGbaSupmayor5 = dataCapGbamayoresA1000.loc[dataCapGbamayoresA1000['surface_total_in_m2'] > 5, :]
dataCapGbaSupmayor5.describe()
```

```python
#Top 5 de barrios con mayor cantidad de departamentos en venta
dataDepartamentos = dataCapGbaSupmayor5.loc[dataCapGbaSupmayor5['property_type'] == 'apartment', :]
dataDepartamentos = dataDepartamentos.loc[:, ["place_name", "price_usd_per_m2"]]
departamentos = dataDepartamentos.groupby("place_name").agg([np.size, np.mean])
departamentos.loc[departamentos[('price_usd_per_m2', 'size')] > 5, :].sort_values(('price_usd_per_m2', 'size'), ascending=False).head(5)
```

```python
#comparar precios medios de departamentos vs casas en belgrano para ver si afecta que haya tantos departamentos
dataTipoDeInmuebleBelgrano = dataCapGbaSupmayor5.loc[dataCapGbaSupmayor5['place_name'] == 'Belgrano', :]
dataTipoDeInmuebleBelgrano = dataTipoDeInmuebleBelgrano.loc[:, ["property_type", "price_usd_per_m2"]]
tipoDeInmuebleBelgrano = dataTipoDeInmuebleBelgrano.groupby('property_type').agg([np.size, np.mean])
tipoDeInmuebleBelgrano
```

```python
#comparar precios medios de departamentos vs casas en palermo para ver si afecta que haya tantos departamentos
dataTipoDeInmueblePalermo = dataCapGbaSupmayor5.loc[dataCapGbaSupmayor5['place_name'] == 'Palermo', :]
dataTipoDeInmueblePalermo = dataTipoDeInmueblePalermo.loc[:, ["property_type", "price_usd_per_m2"]]
tipoDeInmueblePalermo = dataTipoDeInmueblePalermo.groupby('property_type').agg([np.size, np.mean])
tipoDeInmueblePalermo
```

```python
#comparar precios medios de departamentos vs casas en palermo para ver si afecta que haya tantos departamentos
dataTipoDeInmuebleNordelta = dataCapGbaSupmayor5.loc[dataCapGbaSupmayor5['place_name'] == 'Nordelta', :]
dataTipoDeInmuebleNordelta = dataTipoDeInmuebleNordelta.loc[:, ["property_type", "price_usd_per_m2"]]
tipoDeInmuebleNordelta = dataTipoDeInmuebleNordelta.groupby('property_type').agg([np.size, np.mean])
tipoDeInmuebleNordelta
```

## Una posible explicacion a la gran diferencia entre el valor del metro cuadrado entre los departamentos y las casas en belgrano es que la gente prioriza la vista, por lo que vivir en una casa rodeada de edificios no es atractivo para el comprador y por eso se reduce el valor promedio del metro cuadrado en las casas.
